```python
import pandas as pd

df = pd.read_csv("C:/Users/AMUTHA/Desktop/iphone-sales-analysis/data/apple_products.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

df.groupby("Product Name")["Sale Price"].mean().plot(kind="bar", figsize=(10,5))
plt.title("Average Price by iPhone Model")
plt.show()
```

```python
df["Discount Percentage"].plot(kind="hist")
plt.title("Discount Distribution")
plt.show()
```
